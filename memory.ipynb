```python
from langchain.memory import ConversationSummaryBufferMemory
from langchain_core.messages import HumanMessage, AIMessage
from langchain_openai import ChatOpenAI

# Initialize ConversationSummaryBufferMemory with a specified number of recent messages to keep
memory = ConversationSummaryBufferMemory(llm=ChatOpenAI(), max_token_limit=100)

# Simulate adding some conversation to the memory
memory.chat_memory.add_user_message("Hello, how are you?")
memory.chat_memory.add_ai_message("I'm fine, thank you. How can I help you?")
memory.chat_memory.add_user_message("Can you tell me a joke?")
memory.chat_memory.add_ai_message("Sure! Why don't skeletons fight each other? Because they don't have the guts!")
memory.chat_memory.add_user_message("That's funny. Can you explain why it's funny?")
memory.chat_memory.add_ai_message("The joke is a play on words. 'Guts' can mean courage, but skeletons literally lack internal organs, making it a pun.")

# Retrieve the current memory state
print("Memory before summarization:")
print(memory.load_memory_variables({})["history"])

# Trigger summarization
memory.save_context(
    {"input": "Tell me another joke."},
    {"output": "Why did the scarecrow win an award? Because he was outstanding in his field!"}
)

# Check the memory state again after summarization
print("\nMemory after summarization:")
print(memory.load_memory_variables({})["history"])
```
